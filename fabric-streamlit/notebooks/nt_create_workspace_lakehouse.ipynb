### Create workspace

```python
payload = {
    "displayName": "ws_streamlit",
    "capacityId": f"{capacity_id}",
    "description": "Created from notebook"
}
```

```python
import requests, json

url = f"https://api.fabric.microsoft.com/v1/workspaces/"

headers = {
    "Authorization": "Bearer " + mssparkutils.credentials.getToken("pbi"),
    "Content-Type": "application/json"
}

body = payload

response = requests.post(url, headers=headers, json=body)
workspace_id = json.loads(response.text)["id"]
```
